```python
## scrape atas do Copom em portugues

df_ata_pt = pd.DataFrame()
df = scrape_ata_pt()
df['LinkPagina'] = 'https://www.bcb.gov.br' + df['LinkPagina']
urls = df['LinkPagina'].tolist()

for url in urls:
    idd = url[-8:]
    url_temp = 'https://www.bcb.gov.br/api/servico/sitebcb/atascopom/principal?filtro=IdentificadorUrl%20eq%20%27' + idd + '%27'
    
    try:
        df_2 = pd.read_json(url_temp)
        df_3 = json_normalize(df_2["conteudo"])
    except:
        pass
    
    df_ata_pt = pd.concat([df_ata_pt,df_3], sort=True)
```

```python
df_ata_pt.to_csv('atas.csv', index=False)
```

```python
## scrape atas(minutes) do Copom em ingles

df_ata_en = pd.DataFrame()
df = scrape_ata_en()
df['LinkPagina'] = 'https://www.bcb.gov.br' + df['LinkPagina']
urls = df['LinkPagina'].tolist()

for url in urls:
    idd = url[-8:]
    url_temp = 'https://www.bcb.gov.br/api/servico/sitebcb/copomminutes/principal?filtro=IdentificadorUrl%20eq%20%27' + idd + '%27'
    
    try:
        df_2 = pd.read_json(url_temp)
        df_3 = json_normalize(df_2["conteudo"])
    except:
        pass
    
    df_ata_en = pd.concat([df_ata_en,df_3], sort=True)
```

```python
df_ata_en.to_csv('minutes.csv', index=False)
```

```python
## scrape comunicados do Copom em portugues

df_comun_pt = pd.DataFrame()

for year in years:
    
    df = scrape_comun_pt(year)
        
    try:
        urls = df['Url'].tolist()
        filenames = df['notaCopom'].astype(int).tolist()
    except:
        break
    
    for url, filename in zip(urls, filenames):
        idd = url[-10:-5]
        url_temp = url_comun_pt_api + idd
        try:
            df_2 = pd.read_json(url_temp)
            df_3 = json_normalize(df_2["conteudo"])
        except:
            pass
        df_comun_pt = pd.concat([df_comun_pt,df_3], sort=True)
```

```python
df_comun_pt.to_csv('comunicados.csv', index=False)
```

```python
## scrape comunicados(statements) do Copom em inglês

df_comun_en = pd.DataFrame()

for year in years:
    
    df = scrape_comun_en(year)
        
    try:
        urls = df['Url'].tolist()
        filenames = df['notaCopom'].astype(int).tolist()
    except:
        break
    
    for url, filename in zip(urls, filenames):
        idd = url[-10:-5]
        url_temp = url_comun_en_api + idd
        df_2 = pd.read_json(url_temp)
        df_3 = json_normalize(df_2["conteudo"])
        df_comun_en = pd.concat([df_comun_en,df_3], sort=True)
```

```python
df_comun_en.to_csv('statements.csv', index=False)
```

```python
## scrape Copom decisions

df_decisions = scrape_decisions()
```
